```
import numpy as np

np.random.seed(210131)

a = [round(element, 2) for element in np.random.normal(size=1000)]
b = [round(element, 2) for element in np.random.normal(size=500)]
```

```
a
```

```
b
```

```
set_a = {element for element in a}
set_b = {element for element in b}
```

```
len(set_a)
```

```
len(set_b)
```

```
set_a.union(set_b)
```

```
set_a.difference(set_b)
```

```
set_b.difference(set_a)
```

```
set_b.symmetric_difference(set_a)
```

```
set_a.symmetric_difference(set_b)
```

```
set_b.intersection(set_a)
```
